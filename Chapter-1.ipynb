# Bab 1: The Machine Learning Landscape

## Apa itu Machine Learning?

**Machine Learning (ML)** adalah bidang studi (ilmu dan seni) yang memberi komputer kemampuan untuk belajar dari data tanpa perlu diprogram secara eksplisit. Fokus utamanya adalah mengembangkan sistem yang dapat secara mandiri mengidentifikasi pola dalam data, belajar dari pola tersebut, dan membuat keputusan atau prediksi dengan intervensi manusia yang minimal.

## Mengapa Menggunakan Machine Learning?

ML sangat berguna untuk memecahkan berbagai masalah, terutama untuk:
* **Masalah yang solusinya memerlukan banyak aturan rumit** jika dikerjakan secara manual (contoh: filter spam email).
* **Masalah yang sangat kompleks** di mana pendekatan pemecahan masalah tradisional gagal memberikan solusi yang baik.
* **Lingkungan yang terus berubah**, karena sistem ML dapat beradaptasi dan belajar dari data baru yang masuk secara kontinu.
* **Mendapatkan wawasan dari kumpulan data yang besar dan kompleks**, sebuah proses yang sering disebut sebagai *data mining*.

## Contoh Aplikasi Machine Learning

Berikut adalah beberapa contoh penerapan ML di dunia nyata:
- **Klasifikasi Gambar**: Menggunakan *Convolutional Neural Networks* (CNN) untuk mengenali dan mengklasifikasikan objek di dalam gambar atau video.
- **Deteksi Tumor Medis**: Menggunakan teknik *semantic segmentation* untuk mengidentifikasi dan melokalisasi area tumor pada citra medis.
- **Klasifikasi Teks**: Mengkategorikan artikel berita, sentimen pada ulasan produk, atau menyortir komentar di forum secara otomatis.
- **Pengenalan Suara**: Membangun aplikasi yang dapat diperintah melalui suara, seperti asisten virtual.
- **Memprediksi Nilai (Regresi)**: Memproyeksikan harga rumah di masa depan berdasarkan berbagai faktor seperti lokasi, ukuran, dan data ekonomi.

## Jenis-Jenis Sistem Machine Learning

Sistem ML dapat dikategorikan berdasarkan beberapa kriteria:

### Berdasarkan Supervisi Pelatihan

* ***Supervised Learning*** **(Pembelajaran Terawasi)**: Model dilatih menggunakan data yang sudah memiliki "jawaban" atau label yang benar. Tujuannya adalah agar model belajar memetakan input ke output yang sesuai. Tugas utamanya adalah:
    * **Klasifikasi**: Memprediksi kategori (contoh: mengklasifikasikan email sebagai "spam" atau "bukan spam").
    * **Regresi**: Memprediksi nilai numerik (contoh: memprediksi harga saham).
    * *Algoritma Umum*: Linear Regression, K-Nearest Neighbors (KNN), Decision Trees, Support Vector Machines (SVM).

* ***Unsupervised Learning*** **(Pembelajaran Tidak Terawasi)**: Model dilatih menggunakan data yang tidak memiliki label. Tujuannya adalah untuk menemukan struktur, pola, atau hubungan tersembunyi di dalam data itu sendiri.
    * *Tugas Umum*: Clustering, Anomaly Detection, dan Reduksi Dimensi.
    * *Algoritma Umum*: K-Means, DBSCAN (untuk clustering), Principal Component Analysis (PCA) (untuk reduksi dimensi).

* ***Semisupervised Learning*** **(Pembelajaran Semi-Terawasi)**: Merupakan kombinasi dari dua jenis di atas, di mana model dilatih menggunakan sebagian besar data tidak berlabel dan sebagian kecil data berlabel.

* ***Reinforcement Learning*** **(Pembelajaran Penguatan)**: Sebuah agen (*agent*) belajar dengan cara berinteraksi secara aktif dengan lingkungannya (*environment*). Agen akan menerima imbalan (*reward*) atau hukuman (*punishment*) atas tindakan yang diambilnya, dan dari situ ia belajar untuk menemukan strategi (disebut *policy*) terbaik untuk memaksimalkan total imbalan.

### Berdasarkan Cara Belajar (Online vs. Batch)

* ***Batch Learning***: Model dilatih dengan menggunakan semua data yang tersedia secara sekaligus (dalam satu *batch*). Jika ada data baru, sistem harus dilatih ulang dari awal dengan menggabungkan data lama dan data baru.

* ***Online Learning***: Model dilatih secara bertahap (sekuensial) dengan memasukkan data baru secara satu per satu atau dalam *mini-batch* kecil. Ini memungkinkan sistem untuk beradaptasi dengan cepat terhadap perubahan data dan sangat cocok untuk lingkungan yang dinamis.

### Berdasarkan Cara Generalisasi (Instance vs. Model)

* **Pembelajaran Berdasarkan Instansi (*Instance-Based Learning*)**: Sistem belajar dengan cara menghafal contoh-contoh data latih. Ketika dihadapkan dengan data baru, ia membuat prediksi dengan membandingkan dan mencari kemiripan data baru tersebut dengan data yang sudah dihafal.

* **Pembelajaran Berdasarkan Model (*Model-Based Learning*)**: Sistem belajar dengan cara membangun sebuah model matematis dari data latih. Model ini kemudian digunakan untuk melakukan prediksi pada data baru.

## Tantangan Utama dalam Machine Learning

1.  **Kuantitas Data yang Tidak Cukup**: Sebagian besar algoritma ML memerlukan volume data yang besar agar dapat belajar dan bekerja secara efektif.
2.  **Data Latih yang Tidak Representatif**: Jika data latih tidak mencerminkan distribusi data yang akan dihadapi di dunia nyata, model yang dihasilkan tidak akan bisa melakukan generalisasi dengan baik.
3.  **Data Berkualitas Rendah**: Data di dunia nyata seringkali penuh dengan *error*, *outlier* (pencilan), dan *noise* (derau) yang dapat mengganggu proses belajar dan menurunkan kinerja model.
4.  **Fitur yang Tidak Relevan**: Menggunakan fitur (variabel input) yang tidak memiliki hubungan atau korelasi dengan variabel target dapat membingungkan model dan menghasilkan prediksi yang tidak akurat.
5.  ***Overfitting*** **(Generalisasi Berlebih)**: Terjadi ketika model bekerja sangat baik pada data latih, tetapi kinerjanya sangat buruk pada data baru yang belum pernah dilihat. Ini disebabkan model terlalu kompleks dan mulai "menghafal" noise yang ada pada data latih, bukan mempelajari pola yang sebenarnya.
6.  ***Underfitting*** **(Generalisasi Kurang)**: Terjadi ketika model terlalu sederhana untuk menangkap struktur atau pola yang mendasari data. Akibatnya, model memiliki kinerja yang buruk baik pada data latih maupun data uji.

## Testing dan Validasi

Untuk mengetahui seberapa baik sebuah model dapat melakukan generalisasi pada data baru, data yang ada biasanya dibagi menjadi dua atau tiga set:
* **Training Set**: Set data yang digunakan untuk melatih model.
* **Test Set**: Set data yang "disembunyikan" selama pelatihan dan hanya digunakan pada akhir proses untuk mengevaluasi kinerja akhir model pada data yang belum pernah dilihat. Ini memberikan estimasi seberapa baik model akan bekerja di dunia nyata.
* **Validation Set**: Terkadang, *training set* dibagi lagi untuk membuat *validation set*. Set ini digunakan untuk melakukan *hyperparameter tuning* (menemukan kombinasi parameter terbaik) dan seleksi model tanpa perlu "menyentuh" *test set*.
